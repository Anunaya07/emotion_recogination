```python
import cv2  #importing opencv
import matplotlib.pyplot as plt #importing matplotlib
from deepface import DeepFace  #importing Deepface
import numpy as np #importing Numpy
```

```python
img = cv2.imread('happy_boy.jpg')
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) #BGR
```

```python
predictions = DeepFace.analyze(img)
```

# Drawing a rectangle

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.array(gray,dtype='uint8')
faces = faceCascade.detectMultiScale(gray,1.3,5,0)
for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

# Writing Text On the Image

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, predictions['dominant_emotion'],(0,25),font,1,(255,0,0),2, cv2.LINE_4);
```

```python
import cv2  #importing opencv
import matplotlib.pyplot as plt #importing matplotlib
from deepface import DeepFace  #importing Deepface
import numpy as np #importing Numpy

img = cv2.imread('happy_boy.jpg')
predictions = DeepFace.analyze(img)
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.array(gray,dtype='uint8')
faces = faceCascade.detectMultiScale(gray,1.3,5,0)
for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,255,0), 2)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))


font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img, predictions['dominant_emotion'],(0,25),font,1,(255,0,0),2, cv2.LINE_4);
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
